```python
import astropy.io
from astropy.io import fits
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy
```

```python
data = fits.open('halley_low.fit')
data.info()
```

```python
comet = pd.DataFrame()
comet['Wavelength'] = pd.DataFrame(data[1].data['WAVELENGTH'])
comet['Flux'] = pd.DataFrame(data[1].data['FLUX'])
```

```python
peaks = scipy.signal.find_peaks(comet['Flux'], height=0.00000000000005)
peaks
for peak in peaks[0]:
    plt.axvline(comet['Wavelength'][peak])
plt.plot(comet['Wavelength'], comet['Flux'], color='purple')
plt.xlim([1150, 1350])
```
